```python
palette_dic ={'Closed Lost':red,'Closed Won':green, 'Other':blue}
```

```python
g = df.loc[:,['Total_Taxable_Amount_Currency', 'Stage']]
g['Stage'] = g['Stage'].map({'Closed Won':1,'Closed Lost':0})
g['Won_por_fecha'] = g.groupby('Total_Taxable_Amount_Currency')['Stage'].transform('sum')
g['Lost_por_fecha'] = abs(g.groupby('Total_Taxable_Amount_Currency')['Stage'].transform('count') - g['Won_por_fecha'])
g['Stage_diff'] = abs(g['Won_por_fecha'] - g['Lost_por_fecha'])
g['Stage'] = g['Stage'].map({1:'Exito',0:'Fracaso'})
g = g.sort_values('Stage_diff', ascending=False)

palette_dic ={'Fracaso':red,'Exito':green}
plt.figure(figsize=(12,10))
sns.countplot(y=g['Total_Taxable_Amount_Currency'], hue=g['Stage'], order=g.dropna(subset=['Total_Taxable_Amount_Currency'])\
              .drop_duplicates('Total_Taxable_Amount_Currency').iloc[:15]['Total_Taxable_Amount_Currency'].values, \
             palette=palette_dic)

plt.title('Cantidad de registros en la columna Total_Taxable_Amount_Currency respecto a Stage', fontsize=20);
plt.xlabel('Cantidad de registros', fontsize=16);
plt.ylabel('Tipo de moneda', fontsize=16)
```

```python
#Cambio Total_Taxable_Amount a USD
df.loc[df.Total_Taxable_Amount_Currency == "EUR", "Total_Taxable_Amount"] *= 1.19
df.loc[df.Total_Taxable_Amount_Currency == "JPY", "Total_Taxable_Amount"] *= 0.0096
df.loc[df.Total_Taxable_Amount_Currency == "AUD", "Total_Taxable_Amount"] *= 0.73
df.loc[df.Total_Taxable_Amount_Currency == "GBP", "Total_Taxable_Amount"] *= 1.32

#Renombro la columna ya que ahora todos se muestran en USD
df = df.rename(columns={'Total_Taxable_Amount':'Total_Taxable_Amount_USD'})
```
