```python
start_date = df[["Planned_Delivery_Start_Date","Stage"]].groupby(["Planned_Delivery_Start_Date","Stage"]).agg({"Stage":"count"})
start_date.head(1)
```

```python
start_date.columns={"Opened_Amount"}
start_date.reset_index(inplace=True)
start_date.head(1)
```

```python
aprob_fecha_inicial = start_date.loc[start_date.Stage.str.contains("Closed Won")].sort_values("Opened_Amount",ascending=False)
aprob_fecha_inicial = aprob_fecha_inicial[["Planned_Delivery_Start_Date","Opened_Amount"]]
aprob_fecha_inicial.rename(columns={"Opened_Amount":"Opened_Amount_Won"},inplace=True)
aprob_fecha_inicial.head()
```

```python
aprob_fecha_inicial.plot(x="Planned_Delivery_Start_Date",y="Opened_Amount_Won")

plt.title('Cantidad de casos ganados segÃºn fecha inicial planeada', fontsize=20)
plt.xlabel('Fecha Inicial Planeada', fontsize=16);
plt.ylabel('Casos ganados', fontsize=16)
```

```python
sns.palplot(sns.color_palette('Paired', 10))
```

```python
palette = sns.color_palette('Paired', 10)
red = palette[5]
green = palette[3]
blue = palette[1]
```

```python
palette_dic ={'Closed Lost':red,'Closed Won':green, 'Other':blue}
```

```python
g = df.loc[:,['Total_Taxable_Amount_Currency', 'Stage']]
g['Stage'] = g['Stage'].map({'Closed Won':1,'Closed Lost':0})
g['Won_por_fecha'] = g.groupby('Total_Taxable_Amount_Currency')['Stage'].transform('sum')
g['Lost_por_fecha'] = abs(g.groupby('Total_Taxable_Amount_Currency')['Stage'].transform('count') - g['Won_por_fecha'])
g['Stage_diff'] = abs(g['Won_por_fecha'] - g['Lost_por_fecha'])
g['Stage'] = g['Stage'].map({1:'Exito',0:'Fracaso'})
g = g.sort_values('Stage_diff', ascending=False)

palette_dic ={'Fracaso':red,'Exito':green}
plt.figure(figsize=(12,10))
sns.countplot(y=g['Total_Taxable_Amount_Currency'], hue=g['Stage'], order=g.dropna(subset=['Total_Taxable_Amount_Currency'])\
              .drop_duplicates('Total_Taxable_Amount_Currency').iloc[:15]['Total_Taxable_Amount_Currency'].values, \
             palette=palette_dic)

plt.title('Cantidad de registros en la columna Total_Taxable_Amount_Currency respecto a Stage', fontsize=20);
plt.xlabel('Cantidad de registros', fontsize=16);
plt.ylabel('Tipo de moneda', fontsize=16)
```

```python
g = df.loc[:,['Total_Amount_Currency', 'Stage']]
g['Stage'] = g['Stage'].map({'Closed Won':1,'Closed Lost':0})
g['Won_por_fecha'] = g.groupby('Total_Amount_Currency')['Stage'].transform('sum')
g['Lost_por_fecha'] = abs(g.groupby('Total_Amount_Currency')['Stage'].transform('count') - g['Won_por_fecha'])
g['Stage_diff'] = abs(g['Won_por_fecha'] - g['Lost_por_fecha'])
g['Stage'] = g['Stage'].map({1:'Exito',0:'Fracaso'})
g = g.sort_values('Stage_diff', ascending=False)

palette_dic ={'Fracaso':red,'Exito':green}
plt.figure(figsize=(12,10))
sns.countplot(y=g['Total_Amount_Currency'], hue=g['Stage'], order=g.dropna(subset=['Total_Amount_Currency'])\
              .drop_duplicates('Total_Amount_Currency').iloc[:15]['Total_Amount_Currency'].values, \
             palette=palette_dic)

plt.title('Cantidad de registros en la columna Total_Amount_Currency respecto a Stage', fontsize=20);
plt.xlabel('Cantidad de registros', fontsize=16);
plt.ylabel('Tipo de moneda', fontsize=16)
```
