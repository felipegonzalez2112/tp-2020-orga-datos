```python
palette = sns.color_palette('Paired', 10)
red = palette[5]
green = palette[3]
blue = palette[1]
```

```python
palette_dic ={'Closed Lost':red,'Closed Won':green, 'Other':blue}
```

```python
#El 99% de las oportunidades estan en Closed Won y Closed Lost, las grafico

values = ['Closed Won', 'Closed Lost']
g = df.copy()
g['Percent'] = g['Stage'].map(g['Stage'].value_counts(normalize=True)).mul(100)
g['Stage'] = g['Stage'].replace(['Proposal', 'Negotiation', 'Qualification'], 'Other')
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Stage', y='Percent', data = g, palette=palette_dic)

for p in ax.patches:
    txt = str(p.get_height().round(2)) + '%'
    txt_x = p.get_x() 
    txt_y = p.get_height()
    ax.text(txt_x,txt_y,txt)
```

```python
#O grafico solo para los datos con Closed Lost y Closed Won

values = ['Closed Won', 'Closed Lost']
g = df.loc[df['Stage'].isin(values)].copy()
g['Percent'] = g['Stage'].map(g['Stage'].value_counts(normalize=True)).mul(100)
plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Stage', y='Percent', data = g, palette=palette_dic)

for p in ax.patches:
    txt = str(p.get_height().round(2)) + '%'
    txt_x = p.get_x() 
    txt_y = p.get_height()
    ax.text(txt_x,txt_y,txt)
```

```python
g = df.loc[:,['Product_Family', 'Stage']]
g['Stage'] = g['Stage'].map({'Closed Won':1,'Closed Lost':0})
g['Won_por_family'] = g.groupby('Product_Family')['Stage'].transform('sum')
g['Lost_por_family'] = abs(g.groupby('Product_Family')['Stage'].transform('count') - g['Won_por_family'])
g['Stage_diff'] = abs(g['Won_por_family'] - g['Lost_por_family'])
g['Stage'] = g['Stage'].map({1:'Exito',0:'Fracaso'})
g = g.sort_values('Stage_diff', ascending=False)

palette_dic ={'Fracaso':red,'Exito':green}
plt.figure(figsize=(12,10))
sns.countplot(y=g['Product_Family'], hue=g['Stage'], order=g.dropna(subset=['Product_Family'])\
              .drop_duplicates('Product_Family').iloc[:15]['Product_Family'].values, \
             palette=palette_dic)

plt.title('Las 15 familias de productos con mayor diferencia entre exito y fracaso', size=14, y=1.02,\
             fontweight ='bold')
plt.ylabel('Familia de producto')
plt.xlabel('Cantidad de oportunidades')
plt.legend(title='Resultado de la oportunidad')
```

```python
g = df.loc[:,['Product_Name', 'Stage']]
g['Stage'] = g['Stage'].map({'Closed Won':1,'Closed Lost':0})
g['Won_por_family'] = g.groupby('Product_Name')['Stage'].transform('sum')
g['Lost_por_family'] = abs(g.groupby('Product_Name')['Stage'].transform('count') - g['Won_por_family'])
g['Stage_diff'] = abs(g['Won_por_family'] - g['Lost_por_family'])
g['Stage'] = g['Stage'].map({1:'Exito',0:'Fracaso'})
g = g.sort_values('Stage_diff', ascending=False)

palette_dic ={'Fracaso':red,'Exito':green}
plt.figure(figsize=(12,10))
sns.countplot(y=g['Product_Name'], hue=g['Stage'], order=g.dropna(subset=['Product_Name'])\
              .drop_duplicates('Product_Name').iloc[:15]['Product_Name'].values, \
             palette=palette_dic)

plt.title('Los 15 productos con mayor diferencia entre exito y fracaso', size=14, y=1.02,\
             fontweight ='bold')
plt.ylabel('Nombre de producto')
plt.xlabel('Cantidad de oportunidades')
plt.legend(title='Resultado de la oportunidad')
```
