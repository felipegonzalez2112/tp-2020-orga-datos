```python
#El 99% de las oportunidades estan en Closed Won y Closed Lost, las grafico

values = ['Closed Won', 'Closed Lost']
g = df.copy()
g['Percent'] = g['Stage'].map(g['Stage'].value_counts(normalize=True)).mul(100)
g['Stage'] = g['Stage'].replace(['Proposal', 'Negotiation', 'Qualification'], 'Other')
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Stage', y='Percent', data = g)

for p in ax.patches:
    txt = str(p.get_height().round(2)) + '%'
    txt_x = p.get_x() 
    txt_y = p.get_height()
    ax.text(txt_x,txt_y,txt)
```

```python
#O grafico solo para los datos con Closed Lost y Closed Won

values = ['Closed Won', 'Closed Lost']
g = df.loc[df['Stage'].isin(values)].copy()
g['Percent'] = g['Stage'].map(g['Stage'].value_counts(normalize=True)).mul(100)
plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Stage', y='Percent', data = g)

for p in ax.patches:
    txt = str(p.get_height().round(2)) + '%'
    txt_x = p.get_x() 
    txt_y = p.get_height()
    ax.text(txt_x,txt_y,txt)
```
